Introduction to PANDAS

1. pandas is a python package that provides fast,flexible and expressive data structures.
2. It is one of the libraries for dataframes.
3. It works with "relational" or "labeled" data. # relational means excel.
4. It is the most powerful and flexible open source data analysis/manipulation tool
available in any language.
5. In pandas data structures are in-built.

Main features of PANDAS

1. Missing data:Easy handling of missing data ( representated as "Nan", "NA", or
"NaT").
2. Size mutability: columns can be inserted and deleted from dataframe.
3. Data Alignment: Objects can be explictily aligned to a set of labels.
4. Group by: This feature is to perform split, apply, combine operations on datasets.
5. Slicing, Merging, Indexing, Joining, Reshaping.
